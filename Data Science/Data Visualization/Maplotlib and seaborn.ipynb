## Implementation of Data Visualization using Matplotlib and Seaborn

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

## Scatter Plot

```python
plt.scatter(x,y, marker='*', c='g')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title("Graph in 2D")
```

## Line Chart

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'ro--',linewidth=2, markersize=12)
```

## Save Graph

```python
plt.plot(x,y,'ro--',linewidth=2, markersize=12)
plt.savefig('test.png')
```

## Subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'ro--')
plt.subplot(2,2,3)
plt.plot(x,y,'g+')
plt.subplot(2,2,4)
plt.plot(x,y,'b*')
```

## Pie Chart

```python
labels = ['python','c++','ml','java']
students = [215,130,245,210]
colors = ['yellowgreen','lightcoral','lightskyblue','yellow']

explode = (0.2,0,0,0)

plt.pie(students, labels=labels, colors=colors,autopct='%1.3f%%', explode=explode)
```

```python
labels = ['M3','DBMS','S/W Engg.','Data Science', 'OS']
students = [28,23,25,25,27]
colors = ['yellowgreen','lightcoral','lightskyblue','yellow','pink']

plt.pie(students, labels=labels, colors=colors,autopct='%1.1f%%')
plt.show()
```

# Seaborn Lib

```python
import seaborn as sns

df = sns.load_dataset("tips")
df
```

```python
corr = df.corr()
corr
```

## Heat Map

```python
sns.heatmap(corr)
```

## Joint Plot

```python
df1 = df[['total_bill','tip','size']]
sns.jointplot(data=df1,x='tip',y='total_bill')
```

```python
sns.jointplot(data=df1,x='tip',y='total_bill', kind='reg')
```

## Pair Plot

```python
sns.pairplot(data=df1)
```

```python
sns.pairplot(data=df, hue='sex')
```

## Dist Plot

```python
plt.subplot(3,1,1)
sns.distplot(df['total_bill'])
plt.subplot(3,1,2)
sns.distplot(df['tip'])
plt.subplot(3,1,3)
sns.distplot(df['size'])
```

## Count Plot

```python
sns.countplot(data=df,x='sex')
```

## Boxplot 

```python
x = [i for i in range(1,21)] + [100,200]
print(f'x : {x}')
plt.boxplot(x,vert=False)
plt.show()
```

```python
plt.subplot(3,3,1)
plt.boxplot(df['total_bill'],vert=False)
plt.subplot(3,3,2)
plt.boxplot(df['tip'],vert=False)
plt.subplot(3,3,3)
plt.boxplot(df['size'],vert=False)
plt.show()
```
