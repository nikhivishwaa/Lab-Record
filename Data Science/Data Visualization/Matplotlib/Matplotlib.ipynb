## Implementation of Data Visualization using Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

## Scatter Plot

```python
plt.scatter(x,y, marker='*', c='g')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title("Graph in 2D")
```

## Line Chart

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'ro--',linewidth=2, markersize=12)
```

## Save Graph

```python
plt.plot(x,y,'ro--',linewidth=2, markersize=12)
plt.savefig('test.png')
```

## Subplot

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'ro--')
plt.subplot(2,2,3)
plt.plot(x,y,'g+')
plt.subplot(2,2,4)
plt.plot(x,y,'b*')
```

## Bar Plot

```python
x1 = [2,8,10]
y1 = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x1,y1)
plt.bar(x2,y2)
```

## Histogram

## Pie Chart

```python
labels = ['python','c++','ml','java']
students = [215,130,245,210]
colors = ['yellowgreen','lightcoral','lightskyblue','yellow']

explode = (0.2,0,0,0)

plt.pie(students, labels=labels, colors=colors,autopct='%1.3f%%', explode=explode)
plt.show()
```

```python
labels = ['M3','DBMS','S/W Engg.','Data Science', 'OS']
students = [28,23,25,25,27]
colors = ['yellowgreen','lightcoral','lightskyblue','yellow','pink']

plt.pie(students, labels=labels, colors=colors,autopct='%1.1f%%')
plt.show()
```
