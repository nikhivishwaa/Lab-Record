## Implementation of Data Visualization using Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("tips")
df.head(3)
```

```python
df1 = df[['total_bill','tip','size']]
corr = df1.corr()
corr
```

### Heat Map

```python
plt.figure(figsize=(3,2))
sns.heatmap(corr)
```

### Joint Plot

```python
df1 = df[['total_bill','tip','size']]

sns.jointplot(data=df1,x='tip',y='total_bill')
```

```python
sns.jointplot(data=df1,x='tip',y='total_bill', kind='reg')
```

### Pair Plot

```python
sns.pairplot(data=df1)
```

```python
sns.pairplot(data=df, hue='sex')
```

### Dist Plot

```python
plt.figure(figsize=(16,9))
plt.subplot(3,3,1)
sns.distplot(df['total_bill'])
plt.subplot(3,3,2)
sns.distplot(df['tip'])
plt.subplot(3,3,3)
sns.distplot(df['size'])
```

### Count Plot

```python
sns.countplot(data=df,x='sex')
```

### Boxplot 

```python
x = [i for i in range(1,21)] + [100,200]
print(f'x : {x}')
plt.boxplot(x,vert=False)
plt.show()
```

```python
plt.subplot(3,3,1)
plt.boxplot(df['total_bill'],vert=False)
plt.subplot(3,3,2)
plt.boxplot(df['tip'],vert=False)
plt.subplot(3,3,3)
plt.boxplot(df['size'],vert=False)
plt.show()
```
