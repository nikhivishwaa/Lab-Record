```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the Dataset

```python
df = pd.read_csv('heart.csv')
df
```

### splitting data into x and y

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x.shape, y.shape
```

### split data into train and test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

### Applying Logistic Regresssion model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred[:5]
```

### Evaluating Model

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score
```

```python
y_pred1 = model.predict(x_train)
```

```python
print("train dataset accuracy score:",accuracy_score(y_train,y_pred1))
print("test dataset accuracy score:",accuracy_score(y_test,y_pred))
```

```python
print("train dataset f1 score:",f1_score(y_train,y_pred1))
print("test dataset f1 score:",f1_score(y_test,y_pred))
```

```python
cm = confusion_matrix(y_true=y_train,y_pred=y_pred1)
```

```python
cm
```

```python
pd.DataFrame(cm)
```

```python
sns.heatmap(cm, annot=True, fmt='d', xticklabels=['no disease','disease'], yticklabels=['no disease','disease'])
plt.xlabel('actual value')
plt.ylabel('predicted value')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_train,y_pred1))
```
