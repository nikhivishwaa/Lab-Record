```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, classification_report
```

### Load the Dataset

```python
df = pd.read_csv('iris.csv').iloc[:,1:]
df.head()
```

```python
df.shape
```

### Encoding traget vector

```python
encoder = LabelEncoder()
df['Species'] = encoder.fit_transform(df['Species'])
df.head()
```

```python
encoder.inverse_transform([0,1,2])
```

### Splitting data into x and y

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

### Split data into train and test

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
model = LogisticRegression(solver='sag',multi_class='multinomial')
```

```python
model.fit(x_train,y_train)
```

### Evaluating the model

```python
y_pred = model.predict(x_test)
y_pred1 = model.predict(x_train)
```

```python
print(classification_report(y_train, y_pred1))
```

```python
f1_score(y_train,y_pred1,average='weighted')
```

### visualizing

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(x.values,y.values,model,legend=3)
plt.show()
```
