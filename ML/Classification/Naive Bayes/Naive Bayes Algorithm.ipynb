```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data.keys()
```

```python
features = pd.DataFrame(data['data'], columns=data['feature_names'])
features.head()
```

```python
y = pd.DataFrame(data['target'], columns=['Target'])
y.shape
```

### Creating Dataset

```python
new_data = pd.DataFrame(np.c_[data['data'], data['target']], columns=list(data['feature_names']) + ['Target'])
new_data.head()
```

### Split data into Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features,y['Target'], random_state=12, test_size=0.2)
```

### Gaussian Naive Bias

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model1 = GaussianNB()
model1.fit(x_train, y_train)
```

```python
yhat1 = model1.predict(x_test)
```

### Evaluating model Performance

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model1.score(x_train, y_train)
```

```python
print(classification_report(y_test, yhat1))
```

### Bernoulli Naive Bias

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
model2 = BernoulliNB()
model2.fit(x_train, y_train)
```

```python
yhat2 = model2.predict(x_test)
```

### Evaluating model Performance

```python
model2.score(x_train, y_train)
```

```python
print(classification_report(y_test, yhat2))
```

### Multinomial Naive Bias

```python
from sklearn.naive_bayes import MultinomialNB

model3 = MultinomialNB()
model3.fit(x_train, y_train)
```

```python
yhat3 = model3.predict(x_test)
```

### Evaluating model Performance

```python
model3.score(x_train, y_train)
```

```python
print(classification_report(y_test, yhat3))
```
